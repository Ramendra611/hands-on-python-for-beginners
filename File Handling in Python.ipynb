```python
file = open("message.txt", "r")
info = file.read()
file.close()


print(info)
```

### **Common Modes:**
| Mode   | Description                                         |
| ------ | --------------------------------------------------- |
| `'r'`  | Read mode (default)                                 |
| `'w'`  | Write mode (creates a new file if it doesnâ€™t exist) |
| `'a'`  | Append mode (adds content at the end)               |
| `'r+'` | Read and write mode                                 |
| `'b'`  | Binary mode (used for non-text files like images)   |

## using context manager

```python
file = open("awesomepython.txt", "r")
```

```python
with open("awesomepython.txt", "r") as file:
    info = file.read()
print(info)
```

```python
file.closed
```

```python
with open("awesomepython.txt", "a") as file:
    file.write("11. It is very useful for doing data science related work. ")
# print(info)
```

## Serialisation and Deserialisation
    - Json
    - Pickle
    - joblib

```python
import json
```

```python
list_of_names = ["Ram", "Shyam", "Radheshyam"]

my_dictionary = {
    
    "name":"Ram",
    "city":"Delhi"
}
```

```python
with open("list.json", "w") as file:
    json.dump(list_of_names, file)
```

```python
with open("list.json", "r") as file:
    data = json.load(file)
data
```

```python
type(data)
```

```python
with open("dictionary.json", "w") as file:
    json.dump(my_dictionary, file)
```

```python
with open("dictionary.json", "r") as file:
    data = json.load(file)
```

```python
data
```

```python
my_tuple = (5, 6, 7, 8)
with open("tuple.json", "w") as file:
    json.dump(my_tuple, file)
```

```python
with open("tuple.json", "r") as file:
    data = json.load(file)
```

```python
data =tuple(data)
```

```python
data
```

## pickle

```python
list_of_names = ["Ram", "Shyam", "Radheshyam"]

my_dictionary = {
    
    "name":"Ram",
    "city":"Delhi"
}


my_tuple = (5, 6, 7, 8)
```

```python
import pickle
```

```python
with open("dictionary.pkl", "wb") as file:
    pickle.dump(list_of_names, file)
```

```python
with open("dictionary.pkl" , "rb") as file:
    data = pickle.load(file)
```

```python
data
```

```python
with open("tuple.pkl", "wb") as file:
    pickle.dump(my_tuple, file)
```

```python
with open("tuple.pkl" , "rb") as file:
    data = pickle.load(file)
```

```python
data
```

```python
def print_multiplication_table(number):
    for i in range(1, 10):
        print(f"{number} X {i} = {number * i} ")
    
```

```python
print_multiplication_table(5)
```

```python
with open("multiplication_function.pkl", "wb") as file:
    pickle.dump(print_multiplication_table, file)
```

```python
with open("multiplication_function.pkl", "rb") as file:
    fn = pickle.load(file)
```

```python
fn(9)
```
